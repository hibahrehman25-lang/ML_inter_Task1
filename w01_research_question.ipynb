#LANE_4
This lane focuses on identifying pages that may have lower CTR or engagement despite receiving search visibility. The goal is to support content review decisions using observed search signals.



## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

# Research Question

**Decision:**
Which pages should be reviewed first because they have high visibility but low CTR or weak engagement?

**Action:**
A content reviewer or SEO specialist can improve the page title, meta description, or page content.

**Cost of a Wrong Recommendation:**
If a good page is incorrectly flagged, valuable time may be wasted reviewing it. If a poor-performing page is missed, the opportunity to increase clicks and engagement may be lost.

**Why Machine Learning?**
Machine learning can analyze multiple signals together and rank pages more effectively than simple fixed rules.

# Quick Look at the Data

The starter dataset contains **30,000 content pages** with **44 columns** describing search visibility, CTR, engagement, and content characteristics.

These measurements make Lane 4 suitable because it includes impressions, clicks, CTR, average position, engagement rate, and scroll rate, which are directly related to CTR and engagement opportunities.

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("Dataset Shape:", df.shape)
print("Total Pages:", len(df))

print("Average CTR:", round(df["ctr"].mean(),3))
print("Average Engagement Rate:", round(df["engagement_rate"].mean(),3))
print("Average Scroll Rate:", round(df["scroll_rate"].mean(),3))

print("\nTrend Direction Counts:")
print(df["trend_direction"].value_counts())
```
